```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Dataset .csv", encoding="latin1")
df.head()
```

LEVEL 3-TASK 1 : Restaurant Reviews

```python
df['Review_Length'] = df['Rating text'].astype(str).apply(len)

df[['Review_Length', 'Aggregate rating']].corr()
```

```python
plt.scatter(df['Review_Length'], df['Aggregate rating'], alpha=0.5)
plt.title("Review Length vs Rating")
plt.xlabel("Review Length")
plt.ylabel("Rating")
plt.show()
```

We analyzed the relationship between rating text length and restaurant ratings.
The results show whether longer reviews tend to be associated with higher or lower ratings.
