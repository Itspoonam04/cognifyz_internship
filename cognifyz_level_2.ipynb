```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv", encoding="latin1")
df.head()
```

Level 2 - Task 1 :

Restaurant Rating Analysis

```python
df['Aggregate rating'].plot(kind='hist', bins=10)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
df['Aggregate rating'].value_counts().head()
```

```python
df['Votes'].mean()
```

In this task, we analyzed the distribution of restaurant ratings, identified the most common rating values, and calculated the average number of votes.
This helps in understanding customer rating behavior and engagement levels.

Level 2 - Task 2: Cuisine Combination analysis 

```python
cuisine_ratings = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
cuisine_ratings.head(10)
```

We analyzed different cuisine combinations and compared their average ratings.
This helps in understanding which types of cuisine combinations are more preferred by customers.

Level 2 - Task 3 : Geographic Analysis 

```python
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5)
plt.title("Restaurant Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

We plotted the geographic distribution of restaurants using latitude and longitude.
The plot shows that restaurants are clustered in specific urban areas.

Level 2 - Task 4 : Restaurant Chains 

```python
chain_counts = df['Restaurant Name'].value_counts()
chains = chain_counts[chain_counts > 1]
chains.head()
```

```python
df[df['Restaurant Name'].isin(chains.index)].groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False).head()
```

We identified restaurant chains by finding restaurants that appear multiple times in the dataset.
We then compared their average ratings to understand the performance of popular chains.
