 Level 1 - Tassk 1: Top cuisines

```python
## Top 3 Most Common Cuisines 
```

```python
# Split multiple cuisines into single rows
cuisines = df['Cuisines'].str.split(', ').explode()

# Count top 3 most common cuisines
top3_cuisines = cuisines.value_counts().head(3)

top3_cuisines
```

```python
total_restaurants = df.shape[0]

percentage_top3 = (top3_cuisines / total_restaurants) * 100

percentage_top3
```

##The above analysis shows the top three most common cuisines served by restaurants in the dataset.
We also calculated the percentage of restaurants serving each of these cuisines.
This helps us understand customer preferences and popular food categories

Level 1 - Task 2 : City Analysis

```python
df['City'].value_counts().head(1)
```

```python
avg_city_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
avg_city_rating.head()
```

```python
avg_city_rating.head(1)
```

In this task, we identified the city with the highest number of restaurants and calculated the average ratings for each city.
This helps in understanding which cities perform better in terms of restaurant quality and customer satisfaction

Level 1 - task 3 : Price range distribution

```python
import matplotlib.pyplot as plt

df['Price range'].value_counts().sort_index().plot(kind='bar')
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
(df['Price range'].value_counts(normalize=True) * 100)
```
